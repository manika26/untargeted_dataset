```python
#identifiy the internal standards and storing them in different dataframe named iSTD
istd = ds.loc[ds['Metabolite name'].str.contains('iSTD',regex=False)]
istd.reset_index(drop=True,inplace=True)
istd
```

```python
#storing all column names as list
colname=istd.columns.values.tolist()

#creating list of column names which are not required.
cols_not_req = colname[5:32] + colname[43:]
cols_req=colname[0:5] + colname[32:43]
print(cols_not_req,"\n")
print(cols_req)
```

```python
#eliminating the columns not required.
istd_n = istd.drop(cols_not_req, axis=1)

#converting the datatypes of the columns accordingly
istd_n.infer_objects()
istd_n.dtypes
```

```python
#removing the istd dataframe from the initial dataframe
df=pd.concat([ds,istd]).drop_duplicates(keep=False)
df.infer_objects()
df
```

```python
#creation of new column containg the averages of an iSTD in accordance of samples
istd_n['AVG_iSTD']=istd_n.iloc[:,5:].mean(axis=1).apply(lambda x: '{:.3f}'.format(x)).astype(float)
istd_n
```

```python
#creating columns for AVG_iSTD/Sample

samples=['Biorec009_posHILIC_postFlenniken080', 'Biorec011_posHILIC_postFlenniken100', 'Biorec013_posHILIC_postFlenniken120', 'Flenniken004_posHILIC_230295_118', 'Flenniken027_posHILIC_244901_117', 'Flenniken050_posHILIC_253313_119', 'Flenniken112_posHILIC_230188_116', 'Flenniken199_posHILIC_253314_120', 'Flenniken212_posHILIC_230187_115', 'MtdBlnk018_posHILIC_postFlenniken170','MtdBlnk019_posHILIC_postFlenniken180']
#istd_n.iloc[:,5:] = istd_n.iloc[:,5:].div(istd_n['AVG_iSTD'].values,axis=0)
istd_n[samples] = istd_n[samples].div(istd_n['AVG_iSTD'].values,axis=0)
istd_n.loc['Averages'] = istd_n[samples].mean()
istd_n.drop('AVG_iSTD', inplace=True, axis=1)
istd_n
```

```python
#combining the internal standards and Df to get a dataframe 
data_vertical = pd.concat([istd, df],    # Combine vertically
                          ignore_index = True,
                          sort = False)
data_vertical
```

```python
#to highlight selected data
def highlight_text(val):
    if 'iSTD' in str(val):
        return 'background-color: red'
    elif 'unknown' in str(val):
        return 'background-color: yellow'
    elif 'w/o' in str(val):
        return 'background-color: yellow'
    return ''

data_vertical.style.applymap(highlight_text)
```

```python
#automating for finding the samples by marking the starting and ending points
start= data_vertical.columns.get_loc("MS/MS spectrum")+1
start

end=data_vertical.columns.get_loc("QC")-1
end
```

```python
data_vertical.drop('Sample%RSD', inplace=True, axis=1)
data_vertical.drop('Blank fold change', inplace=True, axis=1)
```

```python
#insert RSD column and insert value

mouse_samples=['Flenniken004_posHILIC_230295_118', 'Flenniken027_posHILIC_244901_117', 'Flenniken050_posHILIC_253313_119', 'Flenniken112_posHILIC_230188_116', 'Flenniken199_posHILIC_253314_120', 'Flenniken212_posHILIC_230187_115']
blanks=['MtdBlnk018_posHILIC_postFlenniken170','MtdBlnk019_posHILIC_postFlenniken180']
data_vertical.insert(5, 'Sample%RSD',"")
data_vertical.insert(6, 'Blank fold change',"")
#data_vertical['Sample%RSD'] = (data_vertical.iloc[:,35:41].std(axis=1)/data_vertical.iloc[:,35:41].mean(axis=1))*100

data_vertical['Sample%RSD']=(data_vertical[mouse_samples].std(axis=1)/data_vertical[mouse_samples].mean(axis=1))*100
data_vertical['Blank fold change']=data_vertical[mouse_samples].max(axis=1)/data_vertical[blanks].mean(axis=1)

data_vertical
```

```python
#saving rows with Blank Fold Change <10 and not including the rows having Metabolite name conintaing istd
index_names = data_vertical[(data_vertical['Blank fold change'] < 10) & (~data_vertical['Metabolite name'].str.contains('iSTD',regex=False))].index
#dropping the indexes with blank folds <10
data_vertical.drop(index_names, inplace = True)

data_vertical
```

```python
#normalize the sample data and stored it as result
row=istd_n.loc['Averages']
normalized = data_vertical[samples].div(row, axis=1)
normalized[samples]
```

```python
#identifying unkowns and w/o
unknown = data_vertical.loc[data_vertical['Metabolite name'].str.contains('|'.join(['Unknown', 'w/o']))]

#removing the unknown dataframe from the initial dataframe
data_new=pd.concat([data_vertical,unknown]).drop_duplicates(keep=False)

#combining the unknowns to get a dataframe 
final_data = pd.concat([data_new, unknown],    # Combine vertically
                          ignore_index = True,
                          sort = False)
final_data
```

```python
with pd.ExcelWriter('Output_new.xlsx') as writer:
    final_data.to_excel(writer,sheet_name='DataCuration')
    #data_vertical.to_excel(writer,sheet_name='DataCuration_old')
    normalized[samples].to_excel(writer,sheet_name='Normalized')
    istd_n.to_excel(writer, sheet_name='ISTD')
    
#df.to_csv("test.csv", sep=' ')
```
